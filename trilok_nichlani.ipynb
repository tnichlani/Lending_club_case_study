# <font color=Green> Recommendations. </font>

```python
# Calculating default rates for different determining factors of loan defaults.

Factors=['annual_inc','dti_bin','emp_length','grade','home_ownership','int_rate_bin','month','year','loan_amnt_bin','pub_rec','purpose','term','verification_status']
df=[]
#Defining Default_Rate function to be used to calculate important deriving factors of defaulters
def Default_Rate(x):
    factor_group=loan.groupby(x)
    factor = factor_group.loan_status_marker.mean().sort_values(ascending=False)
    factor_difference=factor.max()-factor.min()
    return factor_difference
```

```python
# Dataframe to calculate and list down important deriving factors of defaulters
df = pd.DataFrame(columns=['Factors','Default Rate (max difference)'])
for item in Factors:
    new_row = {'Factors':item,'Default Rate (max difference)':Default_Rate(item)}
    df= df.append(new_row, ignore_index=True)

df.sort_values(by='Default Rate (max difference)',ascending=False)
```

<font color=green> Thus, by looking at the above chart and the default rate difference, we can conclude that the top 6 deriving factors are interest_rate, grade, pub_rec, home_ownership, purpose and term. </font>
